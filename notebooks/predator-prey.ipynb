```python
import jax.numpy as jnp
from diffrax import MultiTerm, ODETerm, WeaklyDiagonalControlTerm


def drift(t, y, args):
    a = args  # a is alpha
    assert y.shape == (2,)
    r = y[0]
    f = y[1]
    dr = r * (2.0 - a * f) - 0.5 * (2 + a * f)
    df = f * (a * r - 1) - 0.5 * (a * r + 1)
    return jnp.array([dr, df], dtype=y.dtype)


def diffusion(t, y, args):
    a = args  # a is alpha
    assert y.shape == (2,)
    r = y[0]
    f = y[1]
    dr = jnp.sqrt(r * (2.0 + a * f))
    df = jnp.sqrt(f * (a * r + 1.0))
    return jnp.array([dr, df], dtype=y.dtype)


def get_terms(bm):
    return MultiTerm(ODETerm(drift), WeaklyDiagonalControlTerm(diffusion, bm))
```
